```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
d=pd.read_csv("Salary_Data-1.csv")
d1=d.fillna(0)
m,c=0,0 
year=np.array(d1['YearsExperience'])
salary=np.array(d1['Salary'])
ype=[]
mse=[]
yp=np.array(year)
def ce(m,c): 
  yp=m*year+c
  ype.append(np.square(np.subtract(salary,yp)))
  mse.append(np.mean(ype))
for i in range(0,100):
  ce(m+i,c+i)
a=np.arange(0,100)
pt.plot(a,mse,linestyle="dotted",color="red")
pt.xlabel("M and C values")
pt.ylabel("Mse")
pt.title("mean square error graph")
print("The least mean square error is",min(mse),"and the m and c values are",mse.index(min(mse)))
```
